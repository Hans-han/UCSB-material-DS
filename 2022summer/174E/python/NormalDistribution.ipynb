# Illustration: Density and Cumulative Distribution Function of the Normal Distribution

Load some packages:

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
```

Plot PDF for different parameters $\mu$ and $\sigma^2$:

```python
plt.figure(figsize=(8,5))

x = np.linspace(-5,5,1000)

plt.plot(x, norm.pdf(x, loc = 0, scale = np.sqrt(0.2)), linewidth = 2)
plt.plot(x, norm.pdf(x, loc = 0, scale = np.sqrt(1.0)), linewidth = 2)
plt.plot(x, norm.pdf(x, loc = 0, scale = np.sqrt(5.0)), linewidth = 2)
plt.plot(x, norm.pdf(x, loc = -2, scale = np.sqrt(0.5)), linewidth = 2)

plt.xlabel("$x$")
plt.ylabel("$f(x)$")

plt.legend(["$\mu$ = 0, $\sigma^2$ = 0.2", "$\mu$ = 0, $\sigma^2$ = 1.0", 
            "$\mu$ = 0, $\sigma^2$ = 5.0", "$\mu$ = -2, $\sigma^2$ = 0.5"])

plt.title("Probabilitiy density function of the normal distribution")

#plt.savefig("normal_pdf.pdf")
plt.show()
```

Plot CDF for different parameters $\mu$ and $\sigma^2$:

```python
plt.figure(figsize=(8,5))

x = np.linspace(-5,5,1000)

plt.plot(x, norm.cdf(x, loc = 0, scale = np.sqrt(0.2)), linewidth = 2)
plt.plot(x, norm.cdf(x, loc = 0, scale = np.sqrt(1.0)), linewidth = 2)
plt.plot(x, norm.cdf(x, loc = 0, scale = np.sqrt(5.0)), linewidth = 2)
plt.plot(x, norm.cdf(x, loc = -2, scale = np.sqrt(0.5)), linewidth = 2)

plt.xlabel("$x$")
plt.ylabel("$F(x)$")

plt.legend(["$\mu$ = 0, $\sigma^2$ = 0.2", "$\mu$ = 0, $\sigma^2$ = 1.0", 
            "$\mu$ = 0, $\sigma^2$ = 5.0", "$\mu$ = -2, $\sigma^2$ = 0.5"])

plt.title("Cumulative distribution function of the normal distribution")

#plt.savefig("normal_cdf.pdf")
plt.show()
```
