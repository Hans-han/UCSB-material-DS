# Illustration: Density of the Log-normal distribution

Load some packages:

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import lognorm
```

Set the parameters:

```python
S0 = 40
mu = 0.16
sigma = np.array([0.15, 0.2, 0.3, 0.6])
T = 1/2
```

```python
plt.figure(figsize=(10,5))

x = np.linspace(0,100,1000)

plt.plot(x, lognorm.pdf(x, s = sigma[0] * np.sqrt(T), scale = np.exp(np.log(S0)+(mu-0.5*sigma[0]**2)*T)))
plt.plot(x, lognorm.pdf(x, s = sigma[1] * np.sqrt(T), scale = np.exp(np.log(S0)+(mu-0.5*sigma[1]**2)*T)))
plt.plot(x, lognorm.pdf(x, s = sigma[2] * np.sqrt(T), scale = np.exp(np.log(S0)+(mu-0.5*sigma[2]**2)*T)))
plt.plot(x, lognorm.pdf(x, s = sigma[3] * np.sqrt(T), scale = np.exp(np.log(S0)+(mu-0.5*sigma[3]**2)*T)))

plt.legend(["$\sigma$ = 15% p.a.","$\sigma$ = 20% p.a.","$\sigma$ = 30% p.a.","$\sigma$ = 60% p.a."])
plt.xlabel("stock price $S_T$")

plt.title("Densities of the log-normal distribution")

#plt.savefig("lognormal_densities.pdf")
plt.show()
```
