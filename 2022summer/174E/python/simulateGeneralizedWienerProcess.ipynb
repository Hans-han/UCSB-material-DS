# Illustration: Simulation of Generalized Wiener process 

Load some packages:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Simulate Brownian Motion with drift and diffusion coefficient

Simulating a trajectory $(X_t)_{t \in [0,1]}$ of a Brownian motion with drift $\mu = 0.3$ and diffusion coefficient $\sigma = 1.5$ (see Lecture 15, Definition 14.5) on the interval $[0,T]$.

See also the description on slide 5, Lecture 15. 


```python
def simulateGeneralizedBrownianMotion(numberOfSteps, T, x0, mu, sigma):
    
    increments = mu * T/numberOfSteps + np.sqrt(T/numberOfSteps) * sigma * np.random.normal(0, 1, numberOfSteps) 
    
    return x0 + np.append(0, np.cumsum(increments))
```

Plot a simulated path:

```python
numberOfSteps = 10000

mu = 0.3
sigma = 1.5
x0 = 0

T = 100

GenBM1 = simulateGeneralizedBrownianMotion(numberOfSteps, T, x0, mu, sigma)

plt.figure(figsize=(10,5))

plt.plot(np.linspace(0,T,numberOfSteps+1),GenBM1)
plt.plot(np.linspace(0,T,numberOfSteps+1),x0 + mu * np.linspace(0,T,numberOfSteps+1))

plt.xlabel("time $t$")
plt.ylabel("values $B_t$")
plt.title("Trajectory of a Generalized Wiener Process $(X_t)_{0 \leq t \leq T}$ with $x_0=$" + str(x0) + 
          ", $\mu=$" + str(mu) + ", $\sigma=$" + str(sigma) + ", $T$ = "+str(T))
plt.legend(("generalized Wiener Process", "linear function $x_0 + \mu \cdot t$"))

#plt.savefig("picGenBM.pdf")
plt.show()
```
