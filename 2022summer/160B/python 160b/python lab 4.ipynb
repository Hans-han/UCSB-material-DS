### Brownian Motion with drift
$$X_t=\mu t+\sigma W_t$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def simBmDriftScaling(n, mu, sigma):
    bm = np.cumsum(np.random.normal(0,math.sqrt(1/n),n))
    bm = np.array(bm)
    bm = np.insert(bm, 0, 0, axis=0)
    Scaledbm = [x*sigma for x in bm]
    time = [i / (n) for i in range(0,n+1)]
    DriftScaledbm = Scaledbm + np.multiply(time, mu)
    plt.plot(time, DriftScaledbm,label = "mu = " "{}".format(mu)+","+"sigma=""{}".format(sigma))
```

```python
sigma = [1,5]
for sig in sigma:
    simBmDriftScaling(10000,5,sig)
plt.legend(loc='upper left')
```

```python
mu = [0,10, -10]
for m in mu:
    simBmDriftScaling(10000,m,2)
plt.legend(loc='upper left')
```

### Geometric Brownian Motion
$$S_t=S_0\exp\left(\left(\mu-\frac{\sigma^2}{2}\right)t+\sigma W_t\right)$$

```python
def simGBM(S_0, n, mu, sigma):
    bm=0
    GBM=[S_0]
    for k in range(n):
        bm=bm+np.random.normal(0,math.sqrt(1/n))
        X=S_0*math.e**(((mu-sigma**2/2)*k/n)+sigma*bm)
        GBM.append(X)
    time = [i / (n) for i in range(0,n+1)]
    plt.plot(time, GBM,label = "mu = " "{}".format(mu)+","+"sigma=""{}".format(sigma))
```

```python
mu=[0,1,1.3]
for m in mu:
    simGBM(1,10000,m,1)
plt.legend(loc="upper left")
```

```python
sigma=[1,3]
for sig in sigma:
    simGBM(1,10000,2,sig)
plt.legend(loc = "upper left")
```
